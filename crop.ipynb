```python
import cv2
import detect
import numpy as np
from matplotlib import pyplot as plt
```

```python
filename="dataset/LSA64/005_001_001.mp4"
frame_map = detect.FrameMap(min_detection_confidence=0.7)
frames = frame_map.video2frames(filename, interval=10)
coordinates = frame_map.landmarks2value(*frame_map.frame2landmarks(frames[1]))
```

```python
frames[1].shape, coordinates.shape
```

```python
plt.figure(figsize=(5, 5), dpi=100)
plt.imshow(cv2.resize(frames[1], (600, 400)))
plt.show()
```

```python
left_hand = np.array([d for d in coordinates[33:54]]).T
```

```python
height, width = frames[1].shape[:2]
```

```python
min_x, max_x, min_y, max_y = np.array([width * np.min(left_hand[0]), width * np.max(left_hand[0]), height * np.min(left_hand[1]), height * np.max(left_hand[1])]).astype(int)
```

```python
plt.figure(figsize=(5, 5), dpi=100)
plt.imshow(frames[1][min_y:max_y, min_x:max_x, :])
plt.show()
```
